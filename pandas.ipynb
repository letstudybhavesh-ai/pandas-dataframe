Q1. Creating a DataFrame

Create a Pandas DataFrame using a Python dictionary with the following columns:

Product

Region

Sales

Quantity


Add at least 5 rows of sample data and display the DataFrame.

```python
import pandas as pd

data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'Sales': [1500, 200, 350, 800, 450],
    'Quantity': [10, 50, 30, 15, 25]
}
df= pd.DataFrame(data)
print(df)
```

Perform the following inspection tasks:

Show the structure of the dataset using info().

Display the shape of the dataset.

Q3. Column Selection

Select and display only the following columns:

Product

Q4. Indexing using .loc  Using .loc, perform the following:

Display the row with index 10.

Display the Sales value of the row with index 20


```python
print(df.loc[10]) # Row with index 10
print(df.loc[20, 'Sales']) # Sales value of row with index 20
```

Q5. Indexing using .iloc

Using .iloc, display:

1)The first 6 rows and first 4 columns.

2)The Sales value from the 3rd row.

```python
print(df.iloc[0:6, 0:4]) # First 6 rows, first 4 columns
print(df.iloc[2]['Sales']) # Sales from 3rd row
```

Q6. Row Slicing

Display:

1)The first 10 rows of the dataset.

2)Rows 15 to 25.

```python
print(df.iloc[0:10]) # First 10 rows
print(df.iloc[15:26]) # Rows 15 to 25
```

Q7. Filtering Data

Filter the dataset to display:

1)Rows where Region = "East".

2)Rows where Sales > 1500.

```python
print(df[df['Region'] == 'East'].head()) # Region = East
print(df[df['Sales'] > 1500].head()) # Sales > 1500
```

Q8. Multiple Condition Filtering

Display rows where:

Region is West

Sales is greater than 1200

```python
print(df[(df['Region'] == 'West') & (df['Sales'] > 1200)].head())
```

Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

Mean sales value

Maximum quantity sold

Minimum sales value

```python
print(df.describe())
print("\nMean sales value:", df['Sales'].mean())
print("Maximum quantity sold:", df['Quantity'].max())
print("Minimum sales value:", df['Sales'].min())
```

Q10. Aggregation Analysis

Using the Sales column, calculate:

1)Total sales

2)Average sales

3)Maximum sales

4)Minimum sales

```python
print("Total sales:", df['Sales'].sum())
print("Average sales:", df['Sales'].mean())
print("Maximum sales:", df['Sales'].max())
print("Minimum sales:", df['Sales'].min())
```
